```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import random_split
import matplotlib.pyplot as plt
```

```python
# Download the MNIST dataset using PyTorch's torchvision.datasets module
mnist_train_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transforms.ToTensor())
mnist_test_dataset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transforms.ToTensor())
```

```python
# Download the CIFAR-100 dataset using PyTorch's torchvision.datasets module
cifar100_train_dataset = torchvision.datasets.CIFAR100(root='./data', train=True, download=True, transform=transforms.ToTensor())
cifar100_test_dataset = torchvision.datasets.CIFAR100(root='./data', train=False, download=True, transform=transforms.ToTensor())
```

```python
print(f"MNIST dataset loaded with {len(mnist_train_dataset)} training samples and {len(mnist_test_dataset)} test samples.")
print(f"CIFAR-100 dataset loaded with {len(cifar100_train_dataset)} training samples and {len(cifar100_test_dataset)} test samples.")
```

```python
def dataset_info(dataset, name):
    sample, _ = dataset[0]
    print(f"{name} Dataset: Number of training samples: {len(dataset)} Image dimensions: {sample.shape[1]}x{sample.shape[2]} Number of classes: {len(dataset.classes)}")

dataset_info(mnist_train_dataset, "MNIST")
dataset_info(cifar100_train_dataset, "CIFAR-100")
```

```python
# Splitting MNIST
mnist_train, mnist_val = random_split(mnist_train_dataset, [48000, 12000])

# Splitting CIFAR-100
cifar100_train, cifar100_val = random_split(cifar100_train_dataset, [40000, 10000])
```

```python
print(f"MNIST Training set: {len(mnist_train)} samples")
print(f"MNIST Validation set: {len(mnist_val)} samples")
print(f"CIFAR-100 Training set: {len(cifar100_train)} samples")
print(f"CIFAR-100 Validation set: {len(cifar100_val)} samples")
```

```python
def disp(dataset, n, dataset_name):
    image, _ = dataset[n]
    plt.imshow(image.permute(1, 2, 0) if image.shape[0] == 3 else image.squeeze(), cmap='gray')
    plt.title(f"Displaying image number {n} from {dataset_name}")
    plt.show()
```

```python
# Selecting a sample number for 
sample_number = int(input("Enter a sample number to display: "))
```

```python

disp(mnist_train_dataset, sample_number, "MNIST")
disp(cifar100_train_dataset, sample_number, "CIFAR-100")
```

```python
def display_label(dataset, n, dataset_name):
    _, label = dataset[n]
    print(f"Sample {n} label: '{dataset.classes[label]}' ({dataset_name})")

display_label(mnist_train_dataset, sample_number, "MNIST")
display_label(cifar100_train_dataset, sample_number, "CIFAR-100")
```
